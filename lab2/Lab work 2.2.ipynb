### Завдання 2

Програма приймає на вхід два параметри: стать спортсмена(`sex = (f, m)`) та вік спортсмена(`age = [0,100]`).  Спортсмен-дівчина може брати участь у змаганні, якщо її вік більший 18, спортсмен-хлопець - якщо вік більший 16

Якщо введеніть не коректні дані, виведіть повідомлення про помилку

```python
# Вводимо стать та вік спортсмена.
sex = input('Введіть стать для спортсмена (m), для спортсменки (f) ==> ')
age = float(input('Введіть вік спортсмена/спортсменки ==> '))

# Перевірка на дотримання вікового обмеження щодо участі у змаганнях.
if (sex.lower() == "f" and age > 16) or (sex.lower() == "m" and age > 18):
    print('Ви може брати участь у змаганні')

# У разі введення некоректних значень буде виведено повідомлення про помилку.
elif (sex.lower() not in ('f', 'm')) or age <= 0:
    print('Введено помилкові дані')

# Виводимо повідомлення у разі не додержання вікового обмеження щодо участі у змаганнях.
else:
    print('Ви не може брати участь у змаганні')
```

* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import math
frequency={}
# Генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

# Виводимо мінімальне, максимальне, середнє значення послідовності та медіану.    
print(f'Мінімальне значення - {min(list_with_random_numbers)}')
print(f'Максимальне значення - {max(list_with_random_numbers)}')
print(f'Середне значення - {sum(list_with_random_numbers)/len(list_with_random_numbers)}')
print(f'Медіана - {((sorted(list_with_random_numbers)[500])+(sorted(list_with_random_numbers)[501]))/2}')

# Виводимо середнє квадратичне відхилення
my_sum = 0
for s in list_with_random_numbers:
    my_sum = my_sum + ((s - (sum(list_with_random_numbers)/len(list_with_random_numbers)))**2)
print(f'Середнє квадратичне відхилення - {round(math.sqrt(my_sum/(1000-1)),3)}')

for i in list_with_random_numbers:
     frequency[i] = frequency.get(i, 0) + 1

ListFrequency = list(frequency.keys())

print('')
print('Частотa кожної цифри в послідовності. Сортування по цифрі.')
for k in sorted(ListFrequency):
    print(f'Цифра - {k}. {" "*(5-len(str(k)))} Частота - {frequency[k]}')

print('')
print('Частотa кожної цифри в послідовності. Сортування по частоті.')
SortedDictFrequency = sorted(frequency.items(), key=lambda el: el[1], reverse=True)
for k, v in SortedDictFrequency:
    print(f'Цифра - {k}. {" "*(5-len(str(k)))} Частота - {v}')
```
