3. Вивести середнє значення та відхилення курсу за кожний місяць.

```python
import requests
from datetime import datetime, timedelta

# Модуль визначення курсу за параметрами
def geting_currency_rate():
    HEADERS = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:49.0) Gecko/20100101 Firefox/49.0',
                'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
                'Accept-Language': 'en-US,en;q=0.5',
                'Accept-Encoding': 'gzip, deflate',
                'DNT': '1',
                'Connection': 'keep-alive',
                'Upgrade-Insecure-Requests': '1'
            }

    SITE = "https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?json"
    d = datetime(2020, 1, 1) 
    i = 0
    res = []
    while d.year == 2020:
        curd=d.strftime('%Y%m%d')
        d = (d + timedelta(days=1))
        
        f = requests.get(SITE, headers = HEADERS, params = {'date' : curd, 'valcode': 'USD' })
        
        if f.status_code == 200:
            res.append(f.json()[0])
#             print(f.json())
    return res
# Курси за весь рік поміщую у словник
if __name__ == "__main__":
    db = geting_currency_rate()
    print(db)
```

```python
import math
dict_date_rate={}    
db_for_pickle = []

''' This function returns a list of currency rates of US dollar monthly. 
    It has two parametrs: i - number of month and p - list of tuples with exchange date and rate.
'''

def get_rate_per_month (i, p):
    list_els=[]
    list_of_els=[]
    for elm in dict_date_rate.items():
        if i == int(elm[0][3:5]): 
            list_els.append(elm[1])
    return(list_els)
    
    
if __name__ == "__main__":
    
    # Createing dictionary with exchange date and rate. It using for function 'get_rate_per_month'.
    for el in range(len(cur_rate)):

        rate_month_date = cur_rate[el]['exchangedate']
        rate_month = cur_rate[el]['rate']
        dict_date_rate[rate_month_date] = rate_month

    # The loop iterates over for each of month

    for i in range(1,13):
        
        # Determines month number 
        print(f'Місяці: {i}')
        p = dict_date_rate.items()
        
        # Geting the result of  function 'get_rate_per_month'
        list_of_els = get_rate_per_month(i, p)
        
        # Determines mean of carrency rate for each of month
        mean_= round(sum(list_of_els)/len(list_of_els),4)
        print(f' Cереднє значення курсу за місяць ==> {mean_}') 

        # Determines standard deviation for each of month
        list_sd=[]
        list_sd = [(elm - mean_)**2  for elm in list_of_els]
        sd = round(math.sqrt(sum(list_sd)/(len(list_sd)-1)),4)
        print(f' Стандартне відхилення ==> {sd}')
        
        # Createing list with monthly metrics
        db_for_pickle.append( {  'Month':i, 
                                 'Currency':"USD", 
                                 'Mean per month': mean_, 
                                 'sd per month':sd
                              }
                             )
    
    # displays result
    for qw in db_for_pickle:
        print(qw)
```

### Завдання 2:

Потрібно проаналізувати всі товари на сайті: 
https://smallpacking.agrosem.ua/shop/

Створіть базу даних, що містить таблицю із такими стовбцями: 
    - артикул товару
    - найменування
    - ціна 
    - вага

```python
import requests
from bs4 import BeautifulSoup
HEADERS = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:49.0) Gecko/20100101 Firefox/49.0',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
            'Accept-Language': 'en-US,en;q=0.5',
            'Accept-Encoding': 'gzip, deflate',
            'DNT': '1',
            'Connection': 'keep-alive',
            'Upgrade-Insecure-Requests': '1'
        }
spr = []
res = {}
i = 0

# This loop opens each page on site
while i >= 0:
    i += 1
    SITE_NAME = f"https://smallpacking.agrosem.ua/shop/page/{i}/"

    f = requests.get(SITE_NAME, headers = HEADERS)

    if f.status_code == 200:
        soup = BeautifulSoup(f.text, "lxml")
        
        # This loop opens each container with data on page
        for el in soup.find_all("div", class_="product-from-category-container"):
            
            # Finding links for each container on page
            el_con = el.find("a").get("href")
            # Opening each container
            cont = requests.get(el_con)
            if cont.status_code == 200:
                res = {}
   
                s_p = BeautifulSoup(cont.text, "lxml")
                # Finding all issues from the task.
                res['артикул товару'] = (s_p.find('span', class_='sku').text.split()[1])
                res["найменування"] = (s_p.find_all("h2")[1]).text
                res["ціна"] = float((s_p.find("span", class_="regular-price").text).rstrip(' грн').replace(" ", ""))
                res["вага"] = (s_p.find("tr", class_="woocommerce-product-attributes-item woocommerce-product-attributes-item--attribute_pa_fasovka-dobriv").find("td", class_='woocommerce-product-attributes-item__value').text)

            # Creating date base   
            spr.append(res)

    else:
        break

for r in spr:
    print(r)
```

```python
SQL14 ='''
             SELECT 
                    Products.prod_name,
                    Products.prod_price,
                    "< 5"
             FROM Products
             WHERE Products.prod_price < 5
             
            UNION ALL
            
            SELECT 
                    Products.prod_name,
                    Products.prod_price,
                    ">= 5"
             FROM Products
             WHERE Products.prod_price >= 5
 ''' 

curs.execute(SQL14)

for el in curs.fetchall():
     
    print(el)
```
