```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

Створити візуалізацію, аналогічно рисунку 
 - перші 3 графіки візуалізувати, викорстовуючи значення із df
 - останній це пряма пропорційність

```python
%matplotlib inline
```

```python
sns.set_style('white')
x = df.index
y_a = df['a'].values
y_b = df['b'].values
y_c = df['c'].values
y_d = x
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15,8))
ax1.plot(x,y_a )
ax2.plot(x,y_b )
ax3.plot(x,y_c )
ax4.plot(x,y_d )
```

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`


```python
x = df.index
y_a = df['a'].values
y_b = df['b'].values
y_c = df['c'].values
fig, axes = plt.subplots(figsize = (15,10))
sns.set_style('darkgrid')
axes.plot(x, y_a, '-.', color='red')
axes.plot(x, y_b, color='orange', linewidth = 10)
axes.plot(x, y_c,'o-', color='yellow');
```

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
x = df.index
y_a = df['a'].values
y_b = df['b'].values
y_c = df['c'].values
fig, axes = plt.subplots(3, 1, figsize = (15,10))
sns.set_style('darkgrid')
axes[0].plot(x, y_a, label='Line A')
axes[0].legend(loc = 4)
axes[1].plot(x, y_b, label='Line B')
axes[1].legend(loc = 6)
axes[2].plot(x, y_c)
```

```python
!!!!!!!!!!!!!!!!!!!!
```

Створити візуалізацію, аналогічно рисунку 
- використайте томатний колір та відстань між стовбцями 0.5

```python
x = df.index
y = df['a'].values

fig, axes = plt.subplots(figsize = (15,10))
axes.bar(x,y, color='tomato', label='a')
plt.legend()
sns.set_style('darkgrid');
```

Створити візуалізацію, аналогічно рисунку 
- добавте всі підписи та правильний маркер

```python
x = df.index
y = df['a'].values
fig, axes = plt.subplots(figsize = (20,10), sharex='col')
axes.plot(x,y,'^-')
axes.set_title('This is the Title')
axes.set_xlabel('This is the X Axis', fontsize=22)
axes.set_ylabel('This is the Y Axis')
plt.xticks(x)

sns.set_style('darkgrid');
```
